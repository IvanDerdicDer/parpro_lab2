```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('time_results.csv', header=0, sep='|', encoding='utf-16')
```

```python
df = df.groupby('core_count').mean().reset_index()[['core_count', 'time']]
```

```python
t1 = df['time'][0]
```

```python
df['efficiency'] = t1 / (df['time'] * df['core_count'])
df['speedup'] = df['efficiency'] * df['core_count']
df['ideal_speedup'] = df['core_count']
```

```python
df
```

```python
plt.plot(df['core_count'], df['time'])
plt.xlabel('Number of cores')
plt.ylabel('Time (s)')
plt.title('Time to run')
plt.scatter(df['core_count'], df['time'])
plt.show()
```

```python
plt.plot(df['core_count'], df['speedup'])
plt.plot(df['core_count'], df['ideal_speedup'])
plt.scatter(df['core_count'], df['speedup'])
plt.scatter(df['core_count'], df['ideal_speedup'])
plt.xlabel('Number of cores')
plt.ylabel('Speedup')
plt.title('Speedup')
plt.legend(['Speedup', 'Ideal speedup'])
plt.show()
```

```python
plt.plot(df['core_count'], df['efficiency'])
plt.scatter(df['core_count'], df['efficiency'])
plt.xlabel('Number of cores')
plt.ylabel('Efficiency')
plt.title('Efficiency')
plt.show()
```
